①Log in to Weibo to save JSON (the language bar must be completely switched to English before use)

```python
def read_and_save(urls,end,last_end,first_url_flag):
    for pages in range(1,end):
        if first_url_flag == True:
            page = pages#文件保存名称 = 本次加上一次end-1
        else:
            page = pages+last_end
        pyautogui.hotkey('ctrl', 't')
        time.sleep(2)# 等电脑反应
        url = urls + str(pages)
        print(url)
        pyautogui.typewrite(url)
        time.sleep(1)
        pyautogui.hotkey('enter')
        time.sleep(2)
        pyautogui.hotkey('ctrl', 's')
        time.sleep(1)
        string = str(page) + '.json'
        print(string)
        pyautogui.typewrite(string)
        time.sleep(1)
        pyautogui.hotkey('enter')
        time.sleep(2)
        print("保存完了")
        driver.switch_to.window(driver.window_handles[-1])
        driver.close()
```
