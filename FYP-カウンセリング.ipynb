```python
def write_in_txt(all_title_list,path = './data/xinli_output.txt' ):
    #file_handle = open(path,mode='a+',encoding='utf-8')#a+追加
    file_handle = open(path,mode='w',encoding='utf-8')#写
    file_handle.writelines(all_title_list)#将列表中的字符串写入文件中,不会自动换行
    file_handle.close()
```

```python
import csv
def write_in_txt(file_content,file_label,path):
    datas = zip(file_content,file_label)
    with open(path, 'w',encoding='utf-8') as f:
        #创建csv文件的写入对象
        writer = csv.writer(f)
        #写一行,格式:['第1个单元格数据', '第2个单元格数据']
        #writer.writerow(['content', 'label'])
        #写入多行
        writer.writerows(datas)
def write_in_txt2(file_content,path):
    with open(path, 'a', newline='', encoding='utf-8-sig') as f:
        #创建csv文件的写入对象
        writer = csv.writer(f)
        #写一行,格式:['第1个单元格数据', '第2个单元格数据']
        #writer.writerow(['content', 'label'])
        #写入多行
        writer.writerows(file_content)
```

```python
import random
def random_list(filepath,path = './data/sa_train.csv'):
    x = []
    with open(filepath,'r', newline='',encoding='utf_8_sig') as f:
        read = csv.reader(f)
        for i in read:
            if i != []:
                x.append(i)
    random.shuffle(x)
    print(x)
    
    with open(path, 'w', newline='', encoding='utf-8-sig') as f:
            #创建csv文件的写入对象
            writer = csv.writer(f)
            #写一行,格式:['第1个单元格数据', '第2个单元格数据']
            writer.writerow(['content', 'label'])
            writer.writerows(x)
    
            
random_list('./data/train.csv')
random_list('./data/test.csv','./data/sa_test.csv')
#random_list('./data/dev.txt')
print('d')
```
